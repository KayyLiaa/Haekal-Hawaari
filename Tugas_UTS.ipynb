```python
import numpy as np
import matplotlib.pyplot as plt

# Matriks transisi untuk perubahan kualitas tanaman kopi
C = np.array([
    [0.8, 0.15, 0.05],
    [0.1, 0.7, 0.2],
    [0.0, 0.0, 1.0]
])

# Keadaan awal kualitas kopi (Tinggi, Menengah, Rendah)
C0 = np.array([1.0, 0.0, 0.0])

# Periode simulasi 3 periode: 2022, 2023, 2024
periode = 3
distribusi = [C0]

# Proses simulasi distribusi kualitas kopi selama periode
for _ in range(periode - 1):
    distribusi.append(distribusi[-1] @ C)

# Mengubah list distribusi menjadi array numpy
distribusi = np.array(distribusi)

# Membuat data untuk kategori "Terjual Banyak" (hijau), "Stabil" (kuning), "Penurunan" (merah)
tahun = [2022, 2023, 2024]
x = np.linspace(0, 1, periode)
hijau_melengkung = 0.7 * x**2
kuning_melengkung = 0.3 * x**2
merah_melengkung = 0.1 * (4 * x * (1 - x))

# Menampilkan distribusi akhir (periode terakhir)
print("Distribusi akhir (periode terakhir):", distribusi[-1])

# Hasil distribusi kualitas dari tanaman kopi
plt.figure(figsize=(10, 6))
plt.plot(tahun, hijau_melengkung * 100, label="Terjual Banyak", color="green", marker="o")
plt.plot(tahun, kuning_melengkung * 100, label="Stabil", color="yellow", marker="s")
plt.plot(tahun, merah_melengkung * 100, label="Terjual Sedikit", color="red", marker="^")
plt.title("Simulasi Distribusi Total Kopi Terjual (2022 - 2024)")
plt.xlabel("Tahun")
plt.ylabel("Proporsi Kualitas Kopi yang Terjual (%)")
plt.legend()
plt.grid(True)
plt.show()
```
