1. **Parameter LCG**

```python
# Parameter LCG
a = 10
M = 35
Z0 = 33
N = 20
```

2. **Inisialisasi Harga Pasar Saham

```python
# Inisialisasi harga pasar saham
initial_price = 100
prices = [initial_price]
result = []
```

3. **Algoritma LCG, Simulasi Perubahan Harga Saham, dan Menyimpan Hasil Perhitungan**

```python
# Algoritma LCG
Z = Z0
for i in range(1, N + 1):
    Z_prev = Z
    Z = (a * Z) % M
    U = Z / M

    U_rounded = round(U, 3)

    U_str = f"{U_rounded:.3f}"

    # Simulasi perubahan harga saham
    change_percentage = ((Z % 11) - 5) / 100  # Perubahan antara -5% hingga +5%
    current_price = prices[-1]
    new_price = current_price * (1 + change_percentage)
    new_price = round(new_price, 2)  # Harga dibulatkan menjadi 2 desimal

    prices.append(new_price)

    # Menyimpan hasil perhitungan
    rumus = f"(10*{Z_prev}) mod {M}"
    result.append([i, Z_prev, rumus, Z, change_percentage, U_str, current_price, new_price])
```

4. **Output dan Tabel**

```python
# Output
print(f"{'i':<3} | {'Zi-1 (Bilangan Sebelumnya)':<28} | {'Rumus RNG':<25} | {'Zi (RNG)':<12} | {'% Perubahan':<13} | {'Ui (Bil. Acak Seragam)':<22} | {'Harga Sebelum':<15} | {'Harga Sesudah':<15}")
print("-" * 150)

# Tabel
for row in result:
    print(f"{row[0]:<3} | {row[1]:<28} | {row[2]:<25} | {row[3]:<12} | {row[4]:<13.2%} | {row[5]:<22} | {row[6]:<15.2f} | {row[7]:<15.2f}")
```

5. **Visualisasi Harga Pasar Saham**

```python
# Visualisasi harga pasar saham
import matplotlib.pyplot as plt

plt.plot(prices, marker='o', label='Harga Pasar Saham')
plt.title('Simulasi Pergerakan Harga Pasar Saham')
plt.xlabel('Langkah ke-')
plt.ylabel('Harga Pasar Saham')
plt.grid()
plt.legend()
plt.show()
```
