**Teori Rantai Markov**

Rantai Markov ( Markov Chain ) pertama kali dikemukakan oleh matematikawan berkebangsaan Rusia bernama Andrey Andreevich Markov pada tahun 1906. Ini merupakan model yang menggambarkan peluang perubahan suatu kondisi ( state ) berdasarkan kondisi ( state ) pada proses pengamatan sebelumnya. Penyelesaian rantai Markov menggunakan pendekatan matrik dan probabilitas diagram pohon ( probability tree diagram ). Penerapan dalam kehidupan sehari - hari contohnya adalah perubahan cuaca atau iklim pada suatu daerah tertentu atau kebarakan di hutan dan masih banyak hal yang lainnya.

Dalam rantai Markov, probabilitas transisi menunjukkan kemungkinan perpindahan dari satu keadaan ke keadaan lainnya. Hal ini dapat direpresentasikan dalam bentuk matriks transisi probabilitas.

```python
import numpy as np
from IPython.display import Math
import sympy as sym
```

Selanjutnya, disini tampilkan matriks transisi yang berisi kebakaran hutan.

```python
#Matriks Transisi
A = np.array([
    [0.6, 0.3, 0.1],
    [0.0, 0.4, 0.6],
    [0.0, 0.0, 1.0]
])
```

  1. Hijau ke Hijau = 0.6
  2. Hijau ke Terbakar = 0.3
  3. Hijau ke Hangus = 0.1
  4. Terbakar ke Terbakar = 0.4
  5. Terbakar ke Hangus = 0.6
  6. Hangus ke Hangus = 1.0

```python
print("Matriks A =")
display(Math(sym.latex(sym.Matrix(A))))

# Kondisi awal
v0 = np.array([0.7, 0.2, 0.1])
print("Distribusi Awal (v0) =")
display(Math(sym.latex(sym.Matrix(v0))))
```

Untuk kondisi awalnya disini, menunjukkan bahwa
  1. Hijau = 70% kawasan hutan.
  2. Terbakar = 20% kawasan hutan.
  3. Hangus = 10% kawasan hutan.

```python
# Prediksi periode 1
v1 = np.dot(v0, A)
print("Distribusi Periode 1 (v1) =")
display(Math(sym.latex(sym.Matrix(v1))))
```

Pada periode pertama, kita mengalikan kondisi awal dengan matriks transisi untuk mendapatkan distribusi kondisi hutan setelah 1 periode. Sebagian besar kawasan hutan tetap hijau ( 66% ), terbakar ( 28% ), dan hangus ( 6% ).

```python
# Prediksi periode 2
v2 = np.dot(v1, A)
print("Distribusi Periode 2 (v2) =")
display(Math(sym.latex(sym.Matrix(v2))))
```

Untuk periode ke 2, distribusi pada periode pertama dikalikan kembali dengan matriks. Perubahan kondisi mulai terlihat kawasan hijau menjadi ( 62.4% ), dan hangus ( 10.8% ).

```python
# Prediksi periode 3
v3 = np.dot(v2, A)
print("Distribusi Periode 3 (v3) =")
display(Math(sym.latex(sym.Matrix(v3))))
```

Lalu untuk periode ke 3, distribusi pada kedua di kalikan kembali dengan matriks. Disini kawasan hijau semakin sedikit ( 59.5% ) dan kawasan hangus bertambah signifikan ( 14.8% ).

```python
# Prediksi periode 4
v4 = np.dot(v3, A)
print("Distribusi Periode 4 (v4) =")
display(Math(sym.latex(sym.Matrix(v4))))
```

Untuk periode ke 4, distribusi pada periode ke 3 dikalikan kembali dengan matriks. Kawasan hijau menurun lebih lanjut, sementara kawasan yang sudah hangus ( 17.5% ) terus meningkat.

```python
# Prediksi periode 5
v5 = np.dot(v4, A)
print("Distribusi Periode 5 (v5) =")
display(Math(sym.latex(sym.Matrix(v5))))
```

Pada periode ke 5, distribusi pada periode ke 4 dikalikan lagi dengan matriks. Disini kawasan hijau semakin berkurang, dan kawasan yang hangus semakin mendekati angka stabil ( 19.3% ).

```python
# Prediksi periode 6
v6 = np.dot(v5, A)
print("Distribusi Periode 6 (v6) =")
display(Math(sym.latex(sym.Matrix(v6))))
```

Pada periode ke 6, distribusi periode ke 5 dikalikan kembali dengan matriks. Disini kawasan hijau semakin sedikit menjadi ( 55.2% ), dan kawasan yang hangus terus meningkat ( 20.9% ).

**Kondisi Stabil ( Steady State )**

Ketika distribusi tidak berubah lagi antara periode-periode berikutnya, kita akan mencapai kondisi stabil, di mana komposisi kawasan hijau, terbakar, dan hangus tidak akan berubah lebih lanjut. Ini menunjukkan bahwa pada jangka panjang, 50% kawasan hutan akan tetap hijau, 25% akan terbakar, dan 25% akan hangus.

Berikut Codingan Python yang sudah diselesaikan :

```python
import numpy as np
import matplotlib.pyplot as plt

C = np.array([
    [0.7, 0.3, 0.0],
    [0.0, 0.6, 0.4],
    [0.0, 0.0, 1.0]
])

C0 = np.array([0.9, 0.1, 0.0])

periode = 20
distribusi = [C0]


for _ in range(periode):
    distribusi.append(distribusi[-1] @ C)

distribusi = np.array(distribusi)

print("Distribusi akhir (periode terakhir):", distribusi[-1])

plt.figure(figsize=(10, 6))
plt.plot(distribusi[:, 0], label="Sehat", color="green", marker="o")
plt.plot(distribusi[:, 1], label="Terbakar", color="red", marker="s")
plt.plot(distribusi[:, 2], label="Asap", color="gray", marker="^")
plt.title("Simulasi Kebakaran Hutan")
plt.xlabel("Periode")
plt.ylabel("Proporsi Area")
plt.legend()
plt.grid(True)
plt.show()
```
